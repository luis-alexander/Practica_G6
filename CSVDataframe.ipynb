# DataFrame - Amazon Sample

Este DataFrame se obtiene a partir del archivo amazon_sale_report.csv

## Cargar la data desde el CSV

```python
import pandas as pd


df = pd.read_csv("data/amazon_sale_report.csv")
print(df.info())
```

## 2: Explorar estructura

### Exploración de los datos

Se verifica la estructura general del DataFrame para conocer la cantidad de registros y atributos disponibles.

## 4: Identificar columnas con valores nulos
### Identificación de valores nulos

```python

print("\nColumnas con valores nulos:")
print(df.isnull().any())

```

## 5: Contar valores faltantes en cada columna
### Conteo de valores faltantes

```python
print("\nValores faltantes:")
print(df.isnull().sum())
```

### 6. Promedio, máximo y mínimo

```python
print("\nPromedio Amount:", df['Amount'].mean())

print("Promedio:", df['Amount'].mean())
print("Máximo:", df['Amount'].max())
print("Mínimo:", df['Amount'].min())
```

### 7. Máximo y mínimo agrupado

## Agrupación de datos

```python
# v. Máximo y mínimo agrupado por fecha
print("\nMáximo y mínimo de Amount por Status:")
print(df.groupby('Date')['Amount'].agg(['max', 'min']))
```
