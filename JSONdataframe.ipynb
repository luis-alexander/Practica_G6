# DataFrame - Amazon Sample

Este DataFrame se obtiene a partir del archivo JSON generado desde una muestra del dataset original de ventas de Amazon. Se utilizará para explorar la estructura de los datos y realizar análisis descriptivos básicos.

## Cargar la data desde el JSON

```python
import pandas as pd

df_json = pd.read_json("data/amazon_sample.json")
print("Dimensiones:", df_json.shape)
```

## 4: Identificar columnas con valores nulos

### Identificación de valores nulos

Se identifican las columnas que contienen valores faltantes.

```python
columnas_nulas = df_json.columns[df_json.isnull().any()]

print("Columnas con valores nulos:")
print(columnas_nulas)
```

## 5: Contar valores faltantes en cada columna

### Conteo de valores faltantes

Se contabilizan los valores faltantes presentes en cada columna.

```python
df_json.isnull().sum()
```

### 6. Promedio, máximo y mínimo

```python
df_json.columns.tolist()
```

```python
df_json["Amount"] = pd.to_numeric(
    df_json["Amount"],
    errors="coerce"
)

print("Promedio:", df_json["Amount"].mean())
print("Máximo:", df_json["Amount"].max())
print("Mínimo:", df_json["Amount"].min())
```

### 7. Máximo y mínimo agrupado


## Agrupación de datos

Se obtiene el valor máximo y mínimo de ventas agrupado por categoría y estado del pedido.

```python
df_json.groupby(
    ["Category", "Status"]
)["Amount"].agg(
    ["max", "min"]
)
```
